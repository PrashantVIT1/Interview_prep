```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
def ll(l):
  head = None
  curr = None
  for i in range(len(l)):
    new = Node(l[i])
    if head is None:
      head= new
      curr=head
    else:
      curr.next = new
      curr=curr.next
  return head


def mergeTwoSortedLinkedLists(h1,h2):
  p1=h1
  p2=h2
  h3=Node(-1)
  p3=h3
  while (p1 is not None) and p2 is not None:
    if p1.data<p2.data:
      p3.next = p1
      p1=p1.next
    else:
      p3.next = p2
      p2=p2.next
    p3=p3.next
  if p1 is not None:
    p3.next = p1
  if p2 is not None:
    p3.next = p2
  return h3.next

def findNodeRec(head, n) :
  if head is None:
    return -1
  if head.data==n:
    return 0
  f = findNodeRec(head.next,n)
  if f==-1:
    return -1
  else:
    return f+1
def lent(head):
  curr = head
  l = 0
  while curr is not None:
    curr = curr.next
    l = l + 1
  if head is None:
    return None
  return l
def mergeSort(head) : 
  if head is None:
    return  None
  if lent(head)==1:
    return head
  k=1
  curr = head
  while k!=lent(head)//2:
    curr = curr.next
    k=k+1
  temp = curr.next
  curr.next=None
  left = mergeSort(head)
  right = mergeSort(temp)
  return mergeTwoSortedLinkedLists(left,right)
  
  
  

  


h1=ll([7,5,10,14,10,16,3,16,3,16,0,1,10])
h2=ll([])
l1= mergeSort(h1) 
while l1 is not None:
  print(l1.data)
  l1=l1.next
# print("k")
# while l1 is not None:
#   print(l1.data)
#   l1 = l1.next


```
